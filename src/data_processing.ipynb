```python
import cv2
import os

from google.colab import drive
drive.mount('/content/drive')

file_path = "/content/drive/MyDrive/Colab Notebooks/CV/data/UAV/images"
output_path_groundtruth = "/content/drive/MyDrive/Colab Notebooks/CV/data/UAV/groundtruth"
output_path_haze = "/content/drive/MyDrive/Colab Notebooks/CV/data/UAV/haze"

# Create output directories if not exist
os.makedirs(output_path_groundtruth, exist_ok=True)
os.makedirs(output_path_haze, exist_ok=True)

file_list = os.listdir(file_path)

def break_down_image(image):
    height, width, _ = image.shape
    center_x = width // 2

    half_size = 750

    upper_left = image[:, center_x - half_size:center_x]
    upper_right = image[:, center_x:center_x + half_size]

    return upper_left, upper_right

for file_name in file_list:
    full_path = os.path.join(file_path, file_name)

    if os.path.isfile(full_path):
        image = cv2.imread(full_path)

        height = image.shape[0]
        upper_half = image[0:height // 2, :]
        lower_half = image[height // 2:, :]

        upper_left, upper_right = break_down_image(upper_half)
        lower_left, lower_right = break_down_image(lower_half)

        # Save in groundtruth
        name = file_name.split('.png')[0]
        output_groundtruth_path_left = os.path.join(output_path_groundtruth, f"{name}_left.png")
        output_groundtruth_path_right = os.path.join(output_path_groundtruth, f"{name}_right.png")

        cv2.imwrite(output_groundtruth_path_left, upper_left)
        cv2.imwrite(output_groundtruth_path_right, upper_right)

        # Save in haze
        output_haze_path_left = os.path.join(output_path_haze, f"{name}_left.png")
        output_haze_path_right = os.path.join(output_path_haze, f"{name}_right.png")

        cv2.imwrite(output_haze_path_left, lower_left)
        cv2.imwrite(output_haze_path_right, lower_right)

        print(f"Processed and saved {file_name}")
```
